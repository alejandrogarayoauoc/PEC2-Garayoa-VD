# Histograma

Los datos usados en este notebook provienen de Kaggle, concretamente de [esta página](https://www.kaggle.com/datasets/arashnic/taxi-pricing-with-mobility-analytics), y recogen datos de servicios de taxis de una empresa india.

Queremos representar la distancia que han recorrido los taxis de la empresa, para observar la distribución de esta variable y sacar conclusiones sobre que tipo de viajes tienden más a solicitar los clientes (más largos, cortos...).

### Carga y limpieza (de NANs)

```python
import pandas as pd
data = pd.read_csv("datos/sigma_cabs.csv") 
```

La variable que voy a representar es la distancia recorrida por los taxis, por lo que elimino sus valores perdidos.

```python
distance = data['Trip_Distance'].dropna()
```

### Representación

```python
import matplotlib.pyplot as plt

plt.hist(distance, bins=20)
plt.title('Histograma de distancias recorridas por taxis en servicios')
plt.ylabel('Frecuencia')
plt.xlabel('Distancia recorrida por el taxi')

plt.show()
```
