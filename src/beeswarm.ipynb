# Beeswarm chart

Los datos usados en este notebook provienen de Kaggle, concretamente de [esta página](https://www.kaggle.com/datasets/mrsimple07/student-exam-performance-prediction), y recogen datos del desempeño de estudiantes (en términos de aprobado/suspendido) según sus horas de estudio.

Queremos observar cómo se distribuye la variable que recoge el éxito en un examen según las horas de estudio del mismo.


### Carga de datos y ajuste de variables

```python
import pandas as pd
data = pd.read_csv('datos/student_exam_data.csv')
```

Selecciono las columnas que me interesan y decodifico el 0/1 en términos de suspendido/aprobado:

```python
data = data[['Study Hours', "Pass/Fail"]]
data['Aprobado'] = data['Pass/Fail'].map(lambda x: "Aprobado" if x==1 else "Suspendido")
```

### Representación

Uso seaborn para el beeswarm plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.swarmplot(x=data["Aprobado"],
              y=data["Study Hours"],
              palette='Set1',
              hue=data["Aprobado"],
             )

plt.title('Beeswarm plot de la variable "Aprobados" según las horas de estudio')
plt.xlabel('Nota')
plt.ylabel('Horas de estudio')
plt.show()
```
